## Content
Each row represents a customer, each column contains customer's attributes described on the coumn metadat.

The dataset includes information about:

3. Customer account information-how long they have been a customer, contract, payment method, paperless billing, monthly charges, and total charges.

```python

#
# Change each column to its appriopriate datatype
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
cat_cols = [col for col in data.columns if col not in num_cols+['customerID', 'Churn']]

for col in data.columns:
    if col in ['Churn']:
        data[col] = data[col].astype('category').cat.codes.astype('int')
    elif col in cat_cols:
            data[col] = data[col].astype('category')

# Confirm whether the changes have been applied successfully
data.info()
```

```python
data=data.drop(data.index[[488,753,936,1082,1340,3331,3826,4380,5218,6670,6754]],0)
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'])
```

## Feature Engineering

```python
data.drop_duplicates(subset=data.columns, keep='last', inplace=True)
#
#check how many values are in any variables
for col in cat_cols:
    display(col)
    print(data[col].unique(), '\n')
```

```python
cat_cols
```

```python
# encode frequencies of the categories
for col in cat_cols:
    freq = data[col].value_counts().to_dict()
    data[col + '_freq_encode'] = data[col].map(freq)
#
#encode categorical data
data = pd.get_dummies(data = data, columns = cat_cols, drop_first=True)
```

```python
#checking for outliers using box plot
fig, ax=plt.subplots(len(num_cols), figsize=(16,12))
for i, col_val in enumerate(num_cols):
    sns.boxplot(data[col_val], notch= True, ax=ax[i])
    ax[i].set_title ('box plot-'+ col_val, fontsize= 15)
    ax[i].set_xlabel(col_val, fontsize= 12)
    ax[i].set_ylabel('Number', fontsize= 12)

plt.show()  
```

All numeric variables have been accounted for as to have no outliers, the next step is to sclae the data for normality distribution

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
#
data[['tenure', 'MonthlyCharges', 'TotalCharges']] = scaler.fit_transform(data[['tenure', 'MonthlyCharges', 'TotalCharges']])
```

```python
#check for multicollinearity with variance inflation factor (VIF)
from statsmodels.stats.outliers_influence import variance_inflation_factor
#
# feature variables
X=data.drop(['Churn', 'customerID'], axis=1)
y=data['Churn']
#
#VIF dataframe
vif_df=pd.DataFrame()
vif_df['features']=X.columns
#
#calculating VIF for each variable
vif_df['VIF']=[variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
#
#transpose and make row as header
vif_df=vif_df.T
vif_df, vif_df.columns = vif_df[1: ], vif_df.iloc[0]
#
display(vif_df)
```

```python
#drop highly correlated features
import numpy as np
#
corr_matrix=X.corr().abs()

#select upper tringle of correlation matrix
upper=corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

#find index of feature coluns in the correlation greater than .95
to_drop=[column for column in upper.columns if any(upper[column] > 0.95)]

#drop features
X.drop(X[to_drop], axis=1)
```

```python
from collections import Counter
import matplotlib.pyplot as plt
# summarize distribution
counter = Counter(y)
for k,v in counter.items():
    per = v / len(y) * 100
    print('Class=%d, n=%d (%.3f%%)' % (k, v, per))
#
#plot the distribution
plt.bar(counter.keys(), counter.values())
plt.show()
```

```python
#create a balance within the target variable
#oversample the data to balance the target variable
from imblearn.over_sampling import SMOTE
# transform the dataset
oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
```

```python
data=pd.concat([y, X], axis=1)
#write to sv for variable and model selection 
data.to_csv('clean_data.csv', index=False)
```
